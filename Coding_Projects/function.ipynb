```python
#Generators

def func(n=10):
    for i in range(n):
        if i % 2:
            yield i
        
gen = func()
for i in gen:
    print(i)

```

```python
gen = (i for i in range(10) if i %2)
for i in gen:
    print(i)
```

```python
# iterador de 3 x 3 listas

def func(l, n=3)
for i in range(len(l) // n):
    yield l[i*n : (i+1)*n]
    
l = [1,2,3,4,5,6]
gen = func()

for ix, i in enumerate(gen):
    print(ix, i)    
```

```python
def func(start,end):
    result = []
    for i in range(start, end+1):
        if i % 2 == 0:
            result.append(i*2)
    return result

num = func(1,10000)
print(num)
            
```

```python
def process_even_numbers_yield(start, end):
    for i in range(start, end + 1):
        if i % 2 == 0:
            yield i * 2  # Aplicar una operación específica, por ejemplo, multiplicar por 2

even_numbers_generator = process_even_numbers_yield(1, 10000)
for result in even_numbers_generator:
    print(result)
```
